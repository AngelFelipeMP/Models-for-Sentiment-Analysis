```python
import numpy as np
import pandas as pd
import json
from pprint import pprint
```

### 1) Read the JSON file 

```python
## Test
with open('Headlines_Testdata_withscores.json') as f:
    data_test = json.load(f)
pprint(data_test)
```

```python
## Train
with open('Headline_Trainingdata.json') as f:
    data_train = json.load(f)
pprint(data_train)
```

### 2) Convert the JSON to csv and save 

```python
#Convert the dict list to csv and save it into the file Headlines_test.csv
pd.read_json(json.dumps(data_test)).to_csv('Headlines_test.csv')

#Convert the dict list to csv and save it into the file Headlines_train.csv
pd.read_json(json.dumps(data_train)).to_csv('Headlines_train.csv')
```

### 3) Read the CSV file Test

```python
#file test
df_test = pd.read_csv('Headlines_test.csv')

#train
df_train = pd.read_csv('Headlines_train.csv')
```

### 4) looking into the datasets

```python
#Test
df_test.head()
```

```python
#Train
df_train.head()
```

### 5) Organize the data

#### 5.1) Drop the column "Unnamed: 0"

```python
#Test
df_test = df_test.drop(columns=['Unnamed: 0'])
#Train
df_train = df_train.drop(columns=['Unnamed: 0'])
```

#### 5.2) Set the columm "id" as real_id

```python
#Test
df_test.set_index('id', inplace=True)
#Train
df_train.set_index('id', inplace=True)
```

#### 5.3) Check for missing data

```python
#Test
df_test.dropna(inplace=True)
#Train
df_train.dropna(inplace=True)
```

#### 5.4) Drop the column "UniqueID"

```python
#Test
df_test = df_test.drop(columns=['UniqueID'])
```

#### 5.5) Change column name from "sentiment" to "sentiment_range"

```python
#Train
df_test = df_test.rename(columns = {'sentiment score':'sentiment_range'})
#Test
df_train = df_train.rename(columns = {'sentiment':'sentiment_range'})
```

```python
df_test.head()
```

```python
df_train.head()
```

```python
df_test.head()
```

### 6) New Columns

#### Sentiment_label

```python
#Test
df_test['setiment_label'] = np.where(df_test.sentiment_range > 0., 'pos', 'neg')
#Train
df_train['setiment_label'] = np.where(df_train.sentiment_range > 0., 'pos', 'neg')
```

#### Sentiment_score

```python
#Test
df_test['setiment_score'] = np.where(df_test.sentiment_range > 0., 1, 0)
#Train
df_train['setiment_score'] = np.where(df_train.sentiment_range > 0., 1, 0)
```

### 7) Merge data

```python
df_train_and_test = pd.concat([df_test,df_train])
```

```python
df_train_and_test = df_train_and_test.reset_index()
```
